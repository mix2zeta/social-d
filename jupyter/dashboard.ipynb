```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
from pprint import pprint

host = 'social-d-service'

# Date accept format is
# YYYY-MM-DD, YYYY-M-DD, YYYY-M-D, YYYY/MM/DD, YYYY/M/DD, YYYY/M/D, YYYY.MM.DD, 
# YYYY.M.DD, YYYY.M.D, YYYYMMDD, YYYY-DDDD, YYYYDDDD, YYYY-MM, YYYY/MM, YYYY.MM, YYYY
```

```python
# Get number of daily message in date range

from_date = '2019-01-01'
to_date = '2019-01-10'
resp = requests.get(f'http://{host}/date/{from_date}/{to_date}/message/daily')
pprint(resp.json())
```

```python
# Get to 10 account by query message ()

from_date = '2019-03-01'
to_date = '2019-03-10'
query_string = 'samsung AND s10 OR @Chuuchu69' # can use AND, OR in condition
resp = requests.get(f'http://{host}/date/{from_date}/{to_date}/message/top?q={query_string}')
pprint(resp.json())
```

```python
# Get top message by engagements

from_date = '2019-01-01'
to_date = '2019-01-10'
resp = requests.get(f'http://{host}/date/{from_date}/{to_date}/message/engagement')
pprint(resp.json())
```

```python
# Get wordcloud from date

from_date = '2019-01-01'
to_date = '2019-01-02'
resp = requests.get(f'http://{host}/date/{from_date}/{to_date}/message/wordcloud')
pil_im = Image.open(BytesIO(resp.content))
im_array = np.asarray(pil_im)
plt.imshow(im_array)
plt.show()
```
